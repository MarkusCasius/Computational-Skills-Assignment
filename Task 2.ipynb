```python
# 1.i

import matplotlib.pyplot as plt
import numpy as np

# Initialises and declares x to be ‚àí2ùúã ‚â§ ùë• ‚â§ 2ùúã, incrementing in values of 0.1

x = np.arange(-2*np.pi, 2*np.pi, 0.1)

# The following represents the equation ùë¶ = ‚àö2e^(‚àíx2+2)cos(2ùë• +ùúã/6)

y = (np.sqrt(2)*np.exp((-x**2)+2))*np.cos((2*x)+(np.pi/6))

# Applies the grid to the graph and labels the x and y axis, along with a title for readibility

plt.grid()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line graph to show the relationship between x and y')

# Finally plots the data points and displays them

plt.plot(x, y)
plt.show()
```

```python
# 1.ii

# Initialises and declares t to be 0 ‚â§ ùë• ‚â§ 2ùúã, incrementing in values of 0.1

t = np.arange(0, 2*np.pi, 0.001)

# The following represent the expression; ùë•(ùë°) = sin(ùë°)*(ùëí^(cos(ùë°)) ‚àí 2cos(4ùë°) ‚àí sin^(5(ùë°/12)))

x = (np.sin(t))*(np.exp(np.cos(t))-(2*np.cos(4*t))-((np.sin(t/12))**5))

# This line of code represents the expression; y(ùë°) = cos(ùë°)*(ùëí^(cos(ùë°)) ‚àí 2cos(4ùë°) ‚àí sin^(5(ùë°/12)))

y = np.cos(t)*(np.exp(np.cos(t))-2*np.cos(4*t)-(np.sin(t/12)**5))

# Applies the grid to the graph and labels the x and y axis, along with a title for readibility

plt.grid()
plt.xlabel('X(t)')
plt.ylabel('Y(t)')
plt.title('Line graph to show the relationship between x(t) and y(t)')

# Finally plots the data points and displays them

plt.plot(x, y)
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import scipy
import numpy as np

def getVarLength():
    # 2.ii

    # A do-while loop used to get a integer from the user that is greater than 300. If a undesired input is entered, the loop will continue until it is fulfilled.

    while True:
        try:
            varLength = int(input("Please input a integer number greater than 300 "))
            if varLength > 300:
                break
            print("Value less than 300, please try again")
        except ValueError:
            print("Error, please try again")
    return(varLength)


def getRandSeedNum(myID, varLength):
    # 2.iii

    # First Declaring randSeedNum and initialising it with a value of zero, then using a for loop to add each digit from myID to the new variable

    randSeedNum = 0
    for digits in range(len(myID)):
        randSeedNum += int(myID[digits])
    return(randSeedNum)

def createRandomList(randSeedNum):
    # 2.iv

    # First sets the seed that the random library will use to generates its values. Then declares an empty random List and runs a loop, filling it with the random float values until it is the length that the user desired.

    random.seed(randSeedNum)
    randomList = []
    for rangeOfVar in range(varLength):
        randomInteger = random.randint(1, 100)
        randomList.append(randomInteger)
    return(randomList)

def createHistogram(randomList):
    # 2.v

    # Using pyplot function given by matplotlib, the histogram is created using the randomList variable, along with assigning it appropriate labels before showing it.

    plt.hist(randomList, color='purple', edgecolor='black', linewidth=1)
    plt.xlabel("Value of Data")
    plt.ylabel("Frequency of Integers")
    plt.show()

    # Using the scipy and numpy library, the skew, mean and median of the randomList is calculated and printed along with the diagram

    histogramSkew = scipy.stats.skew(randomList)
    print(f'The skew of the graph is:  {histogramSkew}')
    histogramMedian = np.median(randomList)
    print(f'The median of the graph is:  {histogramMedian}')
    histogramMean = np.mean(randomList)
    print(f'The mean of the graph is:  {histogramMean}')

    # -Commenting on the skew, mean and median-
    # The skew is negative, which shows that the 'tail' of the data is more pronounced on the left side. However, the skew value is quite small, making the skew weak and suggesting that the data is random
    # The mean and median are also quite close to the average of 50, being 49.5 and 52 respectively, also suggesting the random nature of the data.


def __main__():
    # 2.i

    # Declaring the myID variable and initialising it a integer number based on my student ID number. Later lines are calls functions to handle the other aspects of the task, passing the needed data from returned calls into the next.

    myID = '2303380'
    varLength = getVarLength()
    randSeedNum = getRandSeedNum(myID, varLength)
    randomList = createRandomList(randSeedNum)
    createHistogram(randomList)



# Runs the Main Function

if __name__ == "__main__":
    __main__()
```

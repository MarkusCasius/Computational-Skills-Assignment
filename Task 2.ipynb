```python
# 1.i

import matplotlib.pyplot as plt
import numpy as np

# Initialises and declares x to be −2𝜋 ≤ 𝑥 ≤ 2𝜋, incrementing in values of 0.1

x = np.arange(-2*np.pi, 2*np.pi, 0.1)

# The following represents the equation 𝑦 = √2e^(−x2+2)cos(2𝑥 +𝜋/6)

y = (np.sqrt(2)*np.exp((-x**2)+2))*np.cos((2*x)+(np.pi/6))

# Applies the grid to the graph and labels the x and y axis for readibility

plt.grid()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line graph to show the relationship between x and y')

# Finally plots the data points

plt.plot(x, y)
plt.show()
```

```python
#1.ii

# Initialises and declares t to be 0 ≤ 𝑥 ≤ 2𝜋, incrementing in values of 0.1

t = np.arange(0, 2*np.pi, 0.1)

# The following represent the expression; 𝑥(𝑡) = sin(𝑡)*(𝑒^(cos(𝑡)) − 2cos(4𝑡) − sin^(5(𝑡/12)))

x = (np.sin(t)(np.exp(np.cos(t)))-(2*np.cos(t))-((np.sin(t))**5))
# x = (np.sin(t)(np.exp(np.cos(t))-2*np.cos(4*t)-(np.sin(t/12)**5)))

# This line of code represents the expression; y(𝑡) = cos(𝑡)*(𝑒^(cos(𝑡)) − 2cos(4𝑡) − sin^(5(𝑡/12)))

y = np.cos(t)(np.exp(np.cos(t))-2*np.cos(4*t)-(np.sin(t/12)**5))

plt.grid()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line graph to show the relationship between x and y')

# Finally plots the data points

plt.plot(x, y)
plt.show()
```

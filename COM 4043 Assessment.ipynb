# Task 1 – Global Aid (Local vs Cloud Storage) 



## Introduction 

Cloud computing refers to the practice of using data servers hosted on the internet to store, manage and process data compared to using a local server or device. There are several service providers which offer consumption-based models for organisations to integrate and use cloud computing. One of such services is Microsoft Azure cloud, however there are some advantages and disadvantages to consider first before migrating systems from remote servers to cloud storage, especially considering that Global aid handles disaster response, which demands robust systems that can handle the strain. This essay shall critically evaluate the strengths and weaknesses of local and cloud storage, to aid in concluding whether transporting systems to Microsoft Azure cloud is a worthwhile investment, or whether its better to maintain the local datacenters Global Aid already possesses. 

## Scalability & Flexibility 

Global aid handles disaster response, meaning there will be times where traffic spikes and great resource demands during emergencies. Local storage suffers from being difficult to scale, with Global aid having to invest into new hardware and upgrades to data servers as time progresses, in addition to finding locations for these data servers. However, cloud storage has greater flexibility and scalability, as the Microsoft Azure cloud offers a consumption-based model, where you can pay as you go for the storage you require. This allows for sufficient data storage during emergencies, and decreasing costs in times where there aren’t any natural disasters. Other organisations like global aid have chosen cloud servers as “scalability, elasticity, pay-per-use pricing, and economies of scale from large scale operations are the major reasons for the successful and widespread adoption of cloud infrastructures.” (Agrawal et al., 2011) 

## Cost-Effectiveness 

When it comes to cost effectiveness, it is a question of short-term vs long-term costs. Cloud service providers cost less initially, requiring just the subscription to start using the service. Whereas local servers must have hardware be bought which is drastically more expensive initially, but with time, eventually those costs should justify themselves. However, local servers must also be maintained by Global Aid, demanding more funds, and scaling up servers also costs a lot more compared to increasing the subscription with the cloud service provider such as Microsoft Azure Cloud. 

However, something important that should be considered is what is known as an egress and ingress charge when using cloud services. Global Aid likely will need to transfer data to and from different regions, whether it be initially transferring the data from the remote datacentres already established, or during the process of using them afterwards. Microsoft Azure cloud has egress charges for data going out from the data server, which should be considered in budgeting as, although it roughly comes to around £0.008 per gigabytes between availability zones and increasing with the distance between the datacentres. While it seems like a low initial price to pay, there is examples  where organisations don’t account for them and pay a great sum on top of contracts, such as an “audit suggests an increased cloud spend of around $30m a year by 2025, as a result of the egress charges, on top of NASA’s $65m-per-year deal with AWS.” (Sharwood & Editor, 2020) Given the potential large sums of data that Global Aid may be processing during disaster response and prevention, considerations should be put toward negotiating a contract with Microsoft Azure cloud to limit these costs if predictions appear too great. 

## Data Security & Privacy 

Considering the data security of cloud, Microsoft Azure cloud offers software and applications along with the subscription to aid reinforce the data security of the storage, in addition to security experts that can be contacted to aid and address any concers Global Aid has. However, a Cloud database is innately owned by a 3rd party, in this case Microsoft, meaning that it is still potentially accessible by people who work for Microsoft, thus the privacy is of concern. This can be limited by defining the service-level agreement (SLA) and defining the responsibility of the service provider to not access the data outside of necessity. It is also coupled with the data protection act and laws that most countries have, the service provider have legal responsibility to ensure that the data stored is private and secure, and any breach of such is told or face legal action. 

On the other hand, local data storage is owned by its user, thus Global aid directly can manage the flow of data into the datacentre and out of, thus increasing the data privacy as there is no other 3rd party with direct access to it. However, there are those who may attempt to access the data illegally, which require Global Aid to hire security experts and set up applications and software to maintain the security of the data, increasing the costs of an already expensive option of data storage. This, as well as being responsible for abiding by the data protection laws within the region the datacentre is stored and the origin of the data, it is a must to protect and use the data responsibly or face legal action. 

## Resilience & Disaster Recovery 

Cloud computing is maintained remotely by datacenters, in regions typically far from sites of frequent natural disaster which Global aid need to provide relief for. Due to this remoteness, it is unlikely they will be affected by natural disasters, compared to having local datacentres in regions of natural disaster which may get damaged and destroyed. Yet, there is also the possibility that the natural disaster severs the infrastructure that maintains connection to the internet, meaning that access to the cloud may not be possible, but a local datacentre within the region can still be used. 

It is also important to consider the likelihood of cyber-attacks or simply the failure of datacentres and loss of data. Microsoft Azure cloud offers inbuilt disaster recovery applications to retrieve data in case of disasters, along with backing up data in different remote datacentres in case of disaster. Local datacentres need to have these applications installed and offsite storage be constructed, which adds to expenses, but don’t innately require connection to the internet and the public knowledge of their existence compared to the popular Microsoft Azure cloud that reduces the potential threat of a cyber-attack. 

## Real-Time Data Processing 

On the matter of data processing, Local datacentres and storage have the upper edge in data transfer as physical connections can offer greater upload speeds compared to cloud computing, which is limited by the bandwidth, and connection to the internet. Of which, Global aid handles disasters like floods and earthquakes, of which those regions may not be fully developed to offer internet connections, making the cloud inaccessible. 

A consideration is that the servers and storage may need to handle big data, which is a term used to describe data transmission which are in high volume (large sums), variety (different data types) and velocity (speed of transmission). Global Aid handles in disaster response, where there is “on average, there are about 6,800 natural disasters that happen every year worldwide.”(The Zebra, 2023) Coupled with it appearing to be on the rise according to “Data captured between 1900 and 2019 by the Institute for Economics and Peace reveal an increase from 39 incidents in 1960 to 396 in 2019.” (Vision of Humanity, 2020), meaning that Global Aid will be met with large demands for data processing that will increase overtime.  

Cloud storage is known for its ability to handle big data due to the interconnectivity of different cloud storages and ease of accessibility, allowing for analysis of the data and acting upon it, as well as the easier scalibility as talked about before in meeting data demands. Whereas remote datacentres will typically contain only information local to the region they are in, making them less efficient in handling big data analysing it. 

## Global Accessibility 

As discussed, prior, cloud storage like Microsoft Azure cloud provide access to the database at any time provided suitable internet connection and bandwidth to receive and send the data. However, remote locations without connection or disasters which damage the infrastructure connecting a region to the internet may lead to an inaccessibility of the data, which is critical considering that Global Aid handles in disaster response and relief and requires constant access to the cloud storage. Another consideration is the downtime of cloud storage due to maintenance and other issues. In the SLA for Microsoft Azure cloud, they least percentage of uptime they guarantee is 99.9%, which roughly equates to 8 hours throughout the year. This downtime should be spread throughout the year, and low time ratio should not be a long-term issue that Global Aid will find to be impeding their disaster relief efforts. 

Local storage however can be accessed both remotely, like cloud storage, but can be accessed physically and through manual physical connections between the datacentre and other site. However, damage to the datacentre due to disasters and being in remote locations may mean the data can't be accessed globally for comparisons and to make decisions. Additionally, remote databases will be inaccessible during maintenance and while upgrading the servers, which can affect accessibility of the storage if there isn’t any backups. Cloud service providers can get around this by duplicating the data and routing clients to a different server and having backups due to a greater investment into their data centres compared to what Global aid would likely be investing if they continued to use their own remote datacentres.  

## Environmental Impact 

According to statistics, “nearly 1% of energy-related global greenhouse gas emissions annually and accounted for approximately 300 metric tons of carbon dioxide equivalent in 2020” (Pratt, 2023). Although this source comes from TechTarget, a profit seeking tech company which may not be credible due to ulterior motives; it is supported by other sources affirming that “The data centres and data transmission networks that underpin digitalisation accounted for around 330 Mt CO2 equivalent in 2020 (including embodied emissions), equivalent to 0.9% of energy-related GHG emissions (or 0.6% of total GHG emissions).” (Rozite et al., 2023) This suggests that datacentres contribute to some of the CO2 emissions to power them, and considerations should be made to limit their impact on the environment.  

While there is always the consideration to continue upgrading the datacentres and making them more efficient, cloud storage innately is more efficient compared to local storage as they are made to support multiple different organisations and companies, which they divide the capabilities of the datacenter between them. Whereas local data centres will only be used by Global Aid, and have reduced efficiency as not all the data storage will be always used as excess will be needed to help support peak traffic and storage. “A 2010 study from Microsoft, Accenture and WSP Environment and Energy found that moving business applications to the cloud could cut the per-user carbon footprint by 30 percent for large, already-efficient companies and as much as 90 percent for the smallest and least efficient businesses.” (Chauhan, 2020) Thus leading to a large incentive to consider moving to cloud storage, to increase efficiency and reduce carbon footprint by collaborating with others.  

## Example of Cloud Computing used in Disaster response 

In 2017, Hurricane Harvey had swept across Texas and Louisiana, a disaster which caused roughly $125 billion in damage and over 100 deaths with the ensuing flooding. In this disaster “The American Red Cross used Amazon Web Services (AWS) to set up a disaster response call center, which handled over 50,000 calls in the first two weeks of the disaster.” During the disaster, the cloud service was able to be used and the datacentre was remote, away from Texas and Louisiana, so there wasn't any issue of it being damaged and data lost. Additionally, the nature of being a cloud service meant that upscaling the cloud service of AWS was a matter of changing the subscription with using a consumption-based model compared to having to upgrade which would’ve happened too late.  

However, Texas was able to maintain its connectivity during the time of crisis, which may not be possible future disasters, which is a struggle and challenge of using cloud servers, and effect even remote servers. It is suggested that “fiber networks—especially the ones built underground—are more durable than copper infrastructure, satellite and wireless.” (Jacobson, 2019) While it may not be within Global Aid’s area of concern, pushing network providers to upgrade their systems in areas of high risk of natural disaster can prevent inaccessibility to both remote and cloud datacentres and aid with disaster response. 

## Conclusion 

Global Aid will find that moving to cloud data storage will yield more advantages compared to local in fields of scalability, security, resilience, accessibility and environmental impact. Microsoft Azure cloud’s service provides improvements in such fields compared to present remote datacentres that Global Aid possess. However, the disadvantages in privacy along with the challenges of utilising cloud datacentres, not just cloud or remote, should be further considered to make the best use of the services provided by the datacentres.

# Analysis of the graph
#By Analysing the graph, the line of best fit indicates a high positive correlation between the opening prices of Apple and Microsoft, with the Spearman's rho correlation coefficient being 0.93, supporting the analysis. Thus, it is safe to conclude that as Apple's opening prices increase, so does Microsoft's opening prices.

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats

# Used to open the csv file to get the Apple and Microsoft prices, and asigns the columns names to later be used for variable assignment.

columns = ['appleOpeningPrice','microsoftOpeningPrice']
incomeData = pd.read_csv('Task4OpenData.csv', usecols=columns)

# Assigns the data to x and y variables to later be used on the graph

x = incomeData.appleOpeningPrice
y = incomeData.microsoftOpeningPrice

# This is used later to draw the line of best fit, using the x and y values assigned previously.

a, b = np.polyfit(incomeData.AppleOpeningPrice, incomeData.MicrosoftOpeningPrice, 1)

# Last of the assignments, res stands for significance result, which is calculated from the square with length equal to the rows in x and y combined. The closer the result is to 1, the stronger the correlation, the greater liklihood for significant results.

res = stats.spearmanr(x, y)

# Increases the size of the displayed graph

plt.figure(figsize=(10,6))

# Layers the graph with a grid for visual aid

plt.grid()

# Increases the size of the displayed plot

# This plots the line of best fit, colouring it blue

plt.plot(incomeData.AppleOpeningPrice, a*incomeData.AppleOpeningPrice+b, color='steelblue')

# This plots the dots on the graph, with a smaller size to improve visual, and coloured purple to make it easier to see compared to 

plt.scatter(x, y, s=10, color='purple')

# These three lines provide the title and labels, x and y respectively, for the graph

plt.title("Scatter Plot: Apple vs MFST Opening Prices with Best Fit Line 2022-2023")
plt.xlabel("Apple Opening Price")
plt.ylabel("Microsoft Opening Price")

# Adds additional text to the graph, showing the equation for the line of best fit, and the Spearman's rho correlation of the graph.

plt.text(132, 350, 'y = ' + '{:.2f}'.format(b) + ' + {:.2f}'.format(a) + 'x', size=14)
plt.text(132, 340, 'R = ' + '{:.2f}'.format(res.statistic), size=14)

# Finally displays the graph

plt.show()
````
